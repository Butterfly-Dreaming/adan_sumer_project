```python
import numpy as np
import cv2 as cv
```

```python
#观测模型
def  观测(H,x):
    return H@x + np.random.normal(0,1/200)
```

```python
#线性模型
def 数学模型(x,T):
    return T@x
```

```python
class 卡尔曼:
    def __init__(self) -> None:
        pass
    def init_Parma(self,A,H,Q,R,P):
        #数学模型
        self.A=A#
     
        #观测
        self.H=H#
        # 协方差
        self.Q=Q#
        self.R=R#
        #P
        self.P=P
    def pridect(self,x,z):
           #先验预测
        self.x_pr = self.A@x 
        print(self.P)
        self.P = self.A@self.P@self.A.T + self.Q
        self.K = self.P@self.H.T/(self.H@self.P@self.H.T+self.R)
        self.x = self.x_pr + self.K@(z - self.H@self.x_pr)
        self.P = self.P - self.K@self.H@self.P
        return self.x
            
       
```

```python
#世界生成；
from cv2 import normalize

po_x_r =0
v_x_r = 3
delt_t = 0.01
t_list = [i for i in range(0,100)]
t_list = np.array(t_list)
t_list =t_list*delt_t
x_list = []
for i in range(0,100):
    x_list.append(t_list[i]*3+np.random.normal(0,1/100))
x_list = np.array(x_list)
x_list
```

```python
#定义有关数学模型的变量
A = np.array([[1,1/100],[0,1]])
X =np.array([[0],[3]])
```

```python
#定义有关观测模型的变量
H = np.array([[1,0]])
观测(H,[x_list[1],3])
```

```python
Q = np.array([ [1/10000,0 ],
               [0,1/10000] ])#数学模型协方差矩阵
R = np.array([[1/40000]])#观测协方差矩阵
P = np.array([[1,0],[0,1]])
```

```python
kammer = 卡尔曼()
kammer.init_Parma(A,H,Q,R,P)
```

```python
x = np.array([[x_list[0]],[3]])
x_kaer= []
x_观测= []
x_数学= []
for i in range(1,100):
    cc = 观测(H,[x_list[i],3])   #直接从传感器中取
    x = kammer.pridect(x,cc)
    x_数学.append(t_list[i]*3)
    x_观测.append(cc[0])
    x_kaer.append(x[0][0])
x_数学=np.array(x_数学)
x_观测=np.array(x_观测)
x_kaer=np.array(x_kaer)
```
